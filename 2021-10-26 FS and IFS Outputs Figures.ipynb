```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Specify the paths to the .csv files with the structural stats
clipIfsPath = "/Users/youngjm/Data/clip/images/derivatives/mpr_ifs_reconall_7.1.1_structural_stats.csv"
qIfsPath = "/Users/youngjm/Data/22q11/derivatives/mpr_ifs_reconall_7.1.1_structural_stats.csv"
clipFsPath = "/Users/youngjm/Data/clip/images/derivatives/mpr_fs_reconall_7.1.1_structural_stats.csv"
qFsPath = "/Users/youngjm/Data/22q11/derivatives/mpr_fs_reconall_7.1.1_structural_stats.csv"

# Load all the files into data frames
clipIfsDf = pd.read_csv(clipIfsPath)
qIfsDf = pd.read_csv(qIfsPath)
clipFsDf = pd.read_csv(clipFsPath)
qFsDf = pd.read_csv(qFsPath)

# Set up colors 
colors = ['xkcd:powder blue', 'xkcd:bright blue', 'xkcd:pale rose', 'xkcd:bright red']
labels = ["IFS CLIP", "FS CLIP", "IFS 22q", "FS 22q"]

# Print the list of column headers
print(list(clipIfsDf))
```

```python
# Plot age vs. gross volume

plt.figure(figsize=(15, 10))

cIfs = plt.scatter(clipIfsDf['age_at_scan_days'], clipIfsDf['EstimatedTotalIntraCranialVol'], c=colors[0], alpha=0.2)
cFs = plt.scatter(clipFsDf['age_at_scan_days'], clipFsDf['EstimatedTotalIntraCranialVol'], c=colors[1], alpha=0.2)
qIfs = plt.scatter(qIfsDf['age_at_scan_days'], qIfsDf['EstimatedTotalIntraCranialVol'], c=colors[2], alpha=0.2)
qFs = plt.scatter(qFsDf['age_at_scan_days'], qFsDf['EstimatedTotalIntraCranialVol'], c=colors[3], alpha=0.2)

plt.title("eTIV vs. Age")
plt.xlabel("Age at Scan (Days)")
plt.ylabel("Estimated Total Intracranial Volums (units)")

plt.legend((cIfs, cFs, qIfs, qFs), labels)
plt.show()
```

```python
# Histogram of Surface Holes aka Euler Numbers

eulerNumbers = [clipIfsDf['SurfaceHoles'].values, clipFsDf['SurfaceHoles'].values, 
                qIfsDf['SurfaceHoles'].values, qFsDf['SurfaceHoles'].values]

plt.figure(figsize=(15, 10))

plt.hist(eulerNumbers, stacked=True, color=colors)

plt.title("Distribution of Surface Holes")
plt.xlabel("Surface Hole Count")
plt.ylabel("Number of Occurrences")

plt.legend(labels)
```

```python
# Histogram of Surface Holes aka Euler Numbers

ages = [clipIfsDf['age_at_scan_days'].values, clipFsDf['age_at_scan_days'].values, 
                qIfsDf['age_at_scan_days'].values, qFsDf['age_at_scan_days'].values]

plt.figure(figsize=(15, 10))

plt.hist(ages, stacked=True, color=colors)

plt.title("Distribution of Age at Scan")
plt.xlabel("Age (Days)")
plt.ylabel("Number of Occurrences")

plt.legend(labels)
```

```python
# Plot age vs. gross volume

plt.figure(figsize=(15, 10))

cIfs = plt.scatter(clipIfsDf['age_at_scan_days'], clipIfsDf['SurfaceHoles'], c=colors[0], alpha=0.5)
cFs = plt.scatter(clipFsDf['age_at_scan_days'], clipFsDf['SurfaceHoles'], c=colors[1], alpha=0.5)
qIfs = plt.scatter(qIfsDf['age_at_scan_days'], qIfsDf['SurfaceHoles'], c=colors[2], alpha=0.5)
qFs = plt.scatter(qFsDf['age_at_scan_days'], qFsDf['SurfaceHoles'], c=colors[3], alpha=0.5)

plt.title("Euler Number vs. Age")
plt.xlabel("Age at Scan (Days)")
plt.ylabel("Number of Holes (units)")

plt.legend((cIfs, cFs, qIfs, qFs), labels)
plt.show()
```
